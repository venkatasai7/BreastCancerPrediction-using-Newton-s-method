```python
def SigmoidLogisticRegression(param):
    a= np.exp(-param)
    a = a+1.0
    sigmoid = np.divide(1.0,a)
    return sigmoid
```

```python
def CalculateCostFunc(Parameters, OutArr):
    d = SigmoidLogisticRegression(Parameters)
    a = np.log(1 - d)
    b = np.log(d)
    c = (1 - OutArr )
    cost = (-OutArr * b - c * a).mean()
    return cost
```

```python
total_iterations = 10
iteration =1
while(iteration<=total_iterations):
    InputArray=np.array(Input)
    OutputArray=np.array(Output)
    OutputArray = OutputArray.reshape(OutputArray.size, 1)
    TrainInput, TestInput, TrainOutput, TestOutput = train_test_split(
                                                                        InputArray, 
                                                                        OutputArray,
                                                                        test_size=0.2,
                                                                        random_state = iteration)
    
    Prameters_at_Training = np.zeros((InputArray.shape[1], 1))
    CalculatedCost = CalculateCostFunc(InputArray @ Prameters_at_Training, OutputArray) 

    TestRowCount = TestInput.shape[0];
    TestRowOutput = TestOutput.shape[0];
    
    ListOfCalculatedCosts = [CalculatedCost]
    InputArrayShape = InputArray.shape[0]
    Classification_Accuracy = []
    TruePositves=0


    while 1:
        SigmoidValue = SigmoidLogisticRegression(InputArray @ Prameters_at_Training) 
        mul = np.multiply(SigmoidValue, (1 - SigmoidValue)) * np.identity(InputArrayShape)
        
        #gradient
        grad_v=(InputArray.T @ (SigmoidValue - OutputArray))/OutputArray.size
        
        
        
        SigmoidValue = (1 / InputArrayShape) *  (InputArray.T @ mul) @ InputArray
        
        Prameters_at_Training= Prameters_at_Training - (np.linalg.inv(SigmoidValue) @ grad_v)
        IterativeCost = CalculateCostFunc(InputArray @ Prameters_at_Training, OutputArray)
        
        ListOfCalculatedCosts.append(IterativeCost)
        
        if math.isnan(IterativeCost)or IterativeCost >= CalculatedCost : 
            break
        else:
            CalculatedCost = IterativeCost     

    ListOfsigmoidValues=[1 if SigmoidLogisticRegression(TestInput[j] @ Prameters_at_Training)>=0.5 else 0 for j in range(TestRowCount)]

    row =0
    while(row<TestRowCount):
        if ListOfsigmoidValues[row] == TestOutput[row]:
            TruePositves += 1
        row+=1
    Classification_Accuracy.append(TruePositves/TestRowOutput)
    print("Accuracy on iteration - ",iteration,": ",TruePositves/TestRowOutput)
    iteration+=1
```

```python
print("Accuracy using Newton's: ", np.mean(Classification_Accuracy))  
```

```python
#inbuilt model in python available in sklearn module
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


total_iterations = 10
iteration =1
Accuracy_Sklearn_LogisticRegression = 0 
while(iteration<=total_iterations):
    Input =input_data.drop(['Diagnosis'],axis=1)
    Output = input_data['Diagnosis'].values
    #dividing the data into 
    TrainInput, TestInput, TrainOutput, TestOutput = train_test_split(
                                                                      Input,Output ,
                                                                      test_size = 0.2,
                                                                      train_size= 0.8 ,
                                                                      shuffle=True,
                                                                      random_state = iteration
                                                                    )
    LinearRegressionModel= LogisticRegression()
    LinearRegressionModel.fit(TrainInput,TrainOutput)
    
    
    print('Score on iteration -\t',iteration,'\t:\t',LinearRegressionModel.score(TestInput,TestOutput))  
    Accuracy_Sklearn_LogisticRegression += LinearRegressionModel.score(TestInput,TestOutput)
    iteration+=1
```
